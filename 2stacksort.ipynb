```python
#THIS CODE WAS GENERATED BY CHATGPT



def sort_with_two_stacks(input_list):
    stack1 = input_list[:]  # Copy of the input list
    stack2 = []  # This will hold the sorted elements

    while stack1:
        # Take the top element from stack1
        temp = stack1.pop()

        # Move elements from stack2 back to stack1 until the right place for temp is found
        while stack2 and stack2[-1] > temp:
            stack1.append(stack2.pop())

        # Place temp in the right spot in stack2
        stack2.append(temp)

    return stack2

# Example usage
unsorted_list = [3, 1, 4, 1, 5, 9, 2, 6, 5]
sorted_list = sort_with_two_stacks(unsorted_list)
print("Sorted list:", sorted_list)
```
